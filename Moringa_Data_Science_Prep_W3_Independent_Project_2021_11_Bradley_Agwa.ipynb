## 1.0 Importing Our Pandas Library

```python
# We will start by importing libraries
import pandas as pd
import numpy as np
```

## 1.1 Loading Our Datasets

```python
# We load our data locally from our csv files.
# Starting with the cells_geo file

cells_geo = pd.read_csv('cells_geo.csv', delimiter = ';')
cells_geo.head()
```

```python
# We will then Load the cells_geo_description dataset
cells_geo_description = pd.read_excel('cells_geo_description.xlsx')
cells_geo_description
```

```python
# From the cells_geo_description we can see the columns that are relevant in the cells_geo dataset.
# Let us select these columns

cells_geo = cells_geo[['VILLES','STATUS', 'LOCALISATION', 'DECOUPZONE', 'ZONENAME', 'REGION','AREA', 'CELL_ID', 'SITE_CODE']]
cells_geo.head()
```

```python
# We will go ahead and load the CDR_Description Dataset

CDR_description = pd.read_excel('CDR_description.xlsx')
CDR_description
```

```python
# We will now load our Telecom Datasets,
# Starting with the first one let's name it day_one

day_one = pd.read_csv('/content/Telcom_dataset.csv')
day_one.head()
```

```python
# Picking out the relevant columns from out dataset above
day_one = day_one[['PRODUTC','VALUE', 'DATETIME', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'CELL_ID','SITE_ID']]
day_one.head()
```

```python
# Renaming the columns in order to have matching column names in each daily dataset
day_one.columns = ['PRODUCT','VALUE', 'DATE_TIME', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'CELL_ID','SITE_ID']
day_one.head()
```

```python
# We proceed to load our second data set, which we will name day_two
day_two = pd.read_csv('/content/Telcom_dataset2.csv')
day_two.head()
```

```python
# Let us get our relevant columns
day_two =day_two[['PRODUCT','VALUE','DATE_TIME','DW_A_NUMBER','DW_B_NUMBER','CELL_ID','SITE_ID']]
day_two.head()
```

```python
# For uniformity we will rename the columns in order to have matching column names in each daily dataset
day_two.columns = ['PRODUCT','VALUE', 'DATE_TIME', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'CELL_ID','SITE_ID']
day_two.head()
```

```python
# On to our final and the third dataset, day_three
day_three = pd.read_csv('/content/Telcom_dataset3.csv')
day_three.head()
```

```python
# Our relevant columns
day_three = day_three[['PRODUCT','VALUE','DATE_TIME','DW_A_NUMBER_INT','DW_B_NUMBER_INT','CELLID','SIET_ID']]
day_three.head()
```

```python
# Just as before we will rename the columns for uniformity with the other datasets
day_three.columns = ['PRODUCT','VALUE', 'DATE_TIME', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'CELL_ID','SITE_ID']
day_three.head()
```

## 1.2 Information On Our Datasets

```python
# Let us check the information of cells_geo dataset by using info()
cells_geo.info()
```

```python
# We will go ahead and replace all the missing values to NaN
cells_geo.replace('N/A',np.NaN)
cells_geo.head()
```

## 1.3 Merging Our Datasets

### Merging Cells_geo with the Day_one Dataset

```python
# We will first merge cells_geo with the day_one dataset.
# Using the inner join we will execute our merge
cell_one = day_one.merge(cells_geo, how= 'inner' , left_on= 'SITE_ID', right_on='SITE_CODE')
cell_one.head()
```

```python
# Converting the cell_one dataframe to a csv file.

cell_one_csv = cell_one.to_csv('cell_one.csv',encoding='utf-8',index=False)

cell_one_csv = pd.read_csv('cell_one.csv')
cell_one_csv.head()
```

```python
# Checking the information details in the newly merged dataset.

cell_one_csv.info()
```

### Merging Cells_geo with the Day_two Dataset

```python
# Moving on we will now merging cells_geo with the day_two dataset.
cell_two = day_two.merge(cells_geo, how= 'inner' , left_on= 'SITE_ID', right_on='SITE_CODE')
cell_two.head()
```

```python
# Just us before ,we will convert the cell_two dataframe to a csv file.

cell_two_csv = cell_two.to_csv('cell_two.csv',encoding='utf-8',index=False)

cell_two_csv = pd.read_csv('cell_two.csv')
cell_two_csv.head()
```

```python
# Checking the information details in the newly merged dataset.

cell_two_csv.info()
```

### Merging Cells_geo with the Day_three Dataset

```python
# Merging our third dataset with the cells_geo
cell_three = day_three.merge(cells_geo, how= 'inner' , left_on= 'SITE_ID', right_on='SITE_CODE')
cell_three.head()
```

```python
# Converting the cell_three dataframe to a csv file to avoid loading the dataset multiple times.

cell_three_csv = cell_three.to_csv('cell_three.csv',encoding='utf-8',index=False)

cell_three_csv = pd.read_csv('cell_three.csv')
cell_three_csv.head()
```

```python
# Checking the information details in the newly merged dataset.

cell_three_csv.info()
```

### Merging all the three Datasets

```python
# Using concatenation we will merge all our three datasets
# We will call our merged set all_com.
all_com = pd.concat([cell_one_csv, cell_two_csv, cell_three_csv], ignore_index=False)
all_com.head()
```

## 1.4 Further Manipulation of Our Data

```python
## Checking for the amount of null values.

all_com.isnull().sum()
```

```python
# Dropping the Status column due to it having a lot of null values
all_com.drop('STATUS', axis = 1, inplace = True)
```

```python
# After which we will run a check to see our change
all_com.isnull().sum()
```

```python
# We will then run a check to confirm our data types
all_com.dtypes
```

```python
# From the info above we can see that our Date_Time needs a change to a datetime type
# Let's do the change


all_com.DATE_TIME = pd.to_datetime(all_com.DATE_TIME)

# Splitting the datetime to date and time


all_com['DATE'] = [d.date() for d in all_com.DATE_TIME]

all_com['TIME'] = [d.time() for d in all_com.DATE_TIME]

all_com['DATE'] = pd.to_datetime(all_com['DATE'])

all_com['TIME'] = pd.to_datetime(all_com['TIME'], format='%H:%M:%S:%f', errors = 'ignore')

# Confirming whether the changes have been made

all_com.dtypes
```

```python
# We can now do a check on our new columns, while also rearraging them.
all_com = all_com[['VILLES','PRODUCT','VALUE','DATE','TIME','DATE_TIME','DW_A_NUMBER_INT','DW_B_NUMBER_INT','LOCALISATION','DECOUPZONE','ZONENAME','REGION','AREA','SITE_ID','SITE_CODE']]
all_com
```
